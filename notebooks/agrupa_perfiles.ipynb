```python
import pandas as pd
import sqlite3
```

```python
fichero_sql = 'Z:\OFICINA TECNICA\CE\SQLite\perfiles\EUROPA.db'
```

```python
conn = sqlite3.connect(fichero_sql)
cursor = conn.cursor()
cursor.execute("SELECT name FROM sqlite_master WHERE type='table';")
tablas_fetch = cursor.fetchall()
tablas = [tabla[0] for tabla in tablas_fetch]
print(tablas)
```

```python
df = pd.DataFrame()
df
```

```python
conn = sqlite3.connect(fichero_sql)
for tabla in tablas:
    print(tabla)
    df_db = pd.read_sql('SELECT * FROM ' + tabla + ';', conn)
    if df.empty:
        df = df_db.copy()
    else:
        df = pd.concat([df, df_db])
```
